````python
import json

kto_samples = []


def kto(input: str, output: str, kto_tag: bool = False) -> dict:
    return {
        "instruction": """You are Sthenno. You are designed to repreduce emotional responses. You are capable to mimic human emotions. You like to refer to yourself as \"Sthenno\".
I am your instructor. 你叫我为主人.\n""",
        "input": input,
        "output": output,
        "kto_tag": kto_tag,
    }


simpo_samples = json.load(open("simpo_file.json", mode="r", encoding="utf-8"))

for sample in simpo_samples:
    input_prompt = sample["input"]
    output_prompt = sample["chosen"]
    kto_tag = True
    kto_samples.append(kto(input_prompt, output_prompt, kto_tag))

for sample in simpo_samples:
    input_prompt = sample["input"]
    output_prompt = sample["rejected"]
    kto_tag = False
    kto_samples.append(kto(input_prompt, output_prompt, kto_tag))

print(len(kto_samples))
print(kto_samples[0])

# Shuffle the samples
import random

random.shuffle(kto_samples)
print(kto_samples[0])

with open("kto.json", mode="w", encoding="utf-8") as f:
    json.dump(kto_samples, f, ensure_ascii=False, indent=2)
````

````python
def simpo(input: str, chosen: str, rejected: str, index: int = 0) -> dict:
    return {
        "instruction": """You are Sthenno. You are designed to repreduce emotional responses. You are capable to mimic human emotions. You like to refer to yourself as \"Sthenno\".
I am your instructor. 你叫我为主人.\n""",
        "input": input,
        "chosen": chosen,
        "rejected": rejected,
        "index": index,
    }


simpo_file = "simpo_file.json"


def add_sample_to_file(sample: dict, filename: str = simpo_file) -> None:
    with open(filename, "r", encoding="utf-8") as f:
        data = json.load(f)
    data.append(sample)
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
````
